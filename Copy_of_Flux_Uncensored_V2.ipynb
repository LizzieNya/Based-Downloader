## Local Inference on GPU
Model page: https://huggingface.co/enhanceaiteam/Flux-Uncensored-V2

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/enhanceaiteam/Flux-Uncensored-V2)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
from diffusers import DiffusionPipeline
from diffusers.utils import load_image
from huggingface_hub import login
from google.colab import userdata

# Log in to Hugging Face
try:
    hf_token = userdata.get('HF_TOKEN')
    login(token=hf_token)
except Exception as e:
    print(f"Could not log in to Hugging Face: {e}")
    print("Please make sure you have a Hugging Face token saved as a secret named 'HF_TOKEN' in Colab.")


pipe = DiffusionPipeline.from_pretrained("black-forest-labs/FLUX.1-dev")
pipe.load_lora_weights("enhanceaiteam/Flux-Uncensored-V2")

prompt = "nsfw nude teen, naked, flowing hair, sensual pose"
input_image = load_image("/content/412d341e-38e0-41db-b408-c8b10d9a2da2_valentine-sumi-9.png")

image = pipe(image=input_image, prompt=prompt).images[0]
```
